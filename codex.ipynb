# Init

```python
import os
import openai
from rich import print
from rich import inspect

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
openai.api_key = os.getenv("OPENAI_API_KEY")


def ai(prompt, model):
    # model = ['text','code']
    response = openai.Completion.create(
        model="{0}-davinci-002".format(model),
        prompt=prompt,
        temperature=0,
        max_tokens=300,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0,
    )

    print(response["choices"][0]["text"])
```

```python
import pandas as pd
from pandas.tseries.offsets import DateOffset
import pandas_datareader.data as web
from datetime import date
import datetime
```

```python
end = date.today()
start= end - DateOffset(months=1) 

end
start
```

```python

df = web.DataReader("SOFI", 'yahoo', start, end)
```

```python
df
```

```python
ai("Use Adj Close from df calculate daily returns",'code')
```

```python
df['daily_returns'] = df['Adj Close'].pct_change()
df
print(f'{df.daily_returns.mean():%}, Average returns for $SOFI in the last month was less than 1% ')
```

```python
dd_path_gen4 = "../data/pl-gen-4/PL Gen4.xlsx"

# load data dict
epn_dd=pd.read_excel(dd_path_gen4, sheet_name="Dict_Experian_Data")

# convert columns to lower case - personal preference > <
epn_dd.columns = [c.lower() for c in epn_dd.columns]
```

```python
# set column names including feature name, type, description, valid values, and default/exclusions
kwargs = {"name_col": "field_name", 
          "type_col": "data_type", 
          "description_col": "description",
          "valid_val_col": "valid values",  # process valid values
          "default_exclusions_col": "default/exclusions"  # process categorical exclusions
         }
dd = dp.ExperianDataDict(epn_dd, **kwargs)
data_dict = dd.data_dict
```
